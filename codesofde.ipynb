```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("CSV Example") \
    .getOrCreate()
```

```python
df_csv = spark.read.csv(r"C:\Users\91838\Downloads\vehicle_sales_data.csv", header=True, inferSchema=True)
df_csv.show()
```

```python
df_csv = spark.read.option("delimiter", ",").csv(r"C:\Users\91838\Downloads\vehicle_sales_data.csv", header=True, inferSchema=True)
df_csv.show()
```

```python
spark = SparkSession.builder \
    .appName("ExcelExample") \
    .config("spark.jars.packages", "com.crealytics:spark-excel_2.13:3.5.1_0.20.4") \
    .getOrCreate()
```

```python
df_excel = spark.read.format("com.crealytics.spark.excel") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(r"C:\Users\91838\Downloads\Vehicle Sales Data.xlsx")
df_excel.show()
```
