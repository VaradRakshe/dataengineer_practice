```python

from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession.builder.appName("RDD Example").getOrCreate()

```

```python
# Get SparkContext
sc = spark.sparkContext

# Create RDD from a Python list
data = [1, 2, 3, 4, 5]
rdd1 = sc.parallelize(data)

# Show elements
print(rdd1.collect())       

```

```python
rdd_from_file = sc.textFile(r"C:\Users\91838\Downloads\abcd.txt")
print(rdd_from_file.take(5))
```

```python
rdd_numbers = sc.parallelize([1, 2, 3, 4, 5])
rdd_squared = rdd_numbers.map(lambda x: x * 2)

print(rdd_squared.collect())
```

```python
rdd_filter = rdd1.filter(lambda x: x % 2 == 0)
print(rdd_filter.collect())
```

```python
rdd_filter = rdd1.filter(lambda x: x==3)
print(rdd_filter.collect())
```
